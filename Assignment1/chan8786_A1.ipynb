Our dataset contains weekly retail prices of three grades of gasoline, diesel, auto propane and compressed natural gas in 10 Ontario markets. It is downloaded from https://data.ontario.ca/dataset/fuels-price-survey-information. Please plot the following visualizations. Your visualization should be complete.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('data/fueltypesall1991-2023.csv')
```

## Question 1

Compare the mean Regular Unleaded Gasoline price among different cities in 2022. Plot the corresponding bar chart.

```python

df_gas2022 = df[(df['Date'] >= '2022-01-01') & (df['Date'] < '2023-01-01') & (df['Fuel Type'] == 'Regular Unleaded Gasoline')]
prices = df_gas2022.select_dtypes(include=[np.number]).mean().drop(['Ontario Average/Moyenne provinciale', 'Southern Average/Moyenne du sud de l\'Ontario', 'Northern Average/Moyenne du nord de l\'Ontario'])

plt.bar(prices.index, prices)
plt.xlabel('Cities')
plt.ylabel('Mean Retail Price')
plt.title('Mean Regular Unleaded Gasoline Price in Different Cities for 2022')
plt.xticks(rotation=90)
plt.show()
```

## Question 2

Plot the histogram of the Regular Unleaded Gasoline price in Ottawa for 2022. Choose the appropriate parameters.

```python
df['Year'] = pd.DatetimeIndex(df['Date']).year
df = df[(df['Year'] == 2022) & (df['Fuel Type'] == 'Regular Unleaded Gasoline')]
plt.hist(df['Ottawa'], bins = 10)
plt.title("Histogram of Regular Unleaded Gasoline Price in Ottawa for 2022")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show
```

## Question 3

Plot the ECDF (normalized) of the Regular Unleaded Gasoline price in Ottawa for 2022.

```python

prices = np.sort(df['Ottawa'])
y = np.arange(1, len(prices) + 1) / len(prices)
plt.plot(prices, y, marker='.', linestyle='none')
plt.xlabel("Prices")
plt.ylabel("ECDF (Normalized)")
plt.title("ECDF of Regular Unleaded Gasoline Price in Ottawa for 2022")
plt.show()
```
